```python
cursor=arcpy.da.UpdateCursor(fc,["UrbRur07","UrbRur17","From07to17"])
urbanized=0
deurbanized=0
for row in cursor:
    if row[0]=="urban" and row[1]=="urban":
        row[2]="no change to be categorized"
    elif row[0]=="rural" and row[1]=="rural":
        row[2]="no change to be categorized"
    elif row[0]=="urban" and row[1]=="rural":
        row[2]="deurbanized"
        deurbanized+=1
    elif row[0]=="rural" and row[1]=="urban":
        row[2]="urbanized"
        urbanized+=1
    cursor.updateRow(row)
del cursor
print "amount urbanized from 2007 to 2017: "+str(urbanized)
print "amount deurbanized from 2007 to 2017: "+str(deurbanized)
    
```
